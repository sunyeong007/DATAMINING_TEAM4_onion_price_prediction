```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
color_pal = sns.color_palette()
plt.style.use('fivethirtyeight')
```

```python
df = pd.read_excel('지수데이터.xlsx')
df
```

```python
# 날짜 범위가 '2018-01-01'에서 '2023-05-17'인 새 데이터 프레임 생성
new_date_range = pd.date_range(start='2013-01-01', end='2023-05-17', name='date_time')
new_withnull_df = pd.DataFrame({'date_time': new_date_range})

# 새 데이터 프레임을 기존 데이터 프레임과 병합
index_df = pd.merge(new_withnull_df, df, on='date_time', how='left')

# date_time을 기준으로 데이터 프레임 정렬
index_df = index_df.sort_values('date_time').reset_index(drop=True)
```

```python
df.shape
```

```python
#### 원본데이터는 월별데이터라 일별데이터로 바꿔줘야 함

# 원하는 너비 및 높이 설정
fig, ax = plt.subplots(figsize=(10, 6))

# plot 표시
ax.plot(df['date_time'], df['total_price_index'], 'ro')
ax.plot(df['date_time'], df['fresh_vegi_index'], 'go')
ax.plot(df['date_time'], df['living_price_index'], 'bo')
ax.plot(df['date_time'], df['food_living_index'], 'yo')

# 플롯 제목 및 범례 설정
ax.set_title('월별 지수')
ax.legend(['total_price_index', 'fresh_vegi_index', 'living_price_index', 'food_living_index'])

# 플롯 표시
plt.show()
```

```python
#### 일별데이터로 바꿔준 모습

# 원하는 너비 및 높이 설정
fig, ax = plt.subplots(figsize=(10, 6))

# plot 표시
ax.plot(new_index_df['date_time'], new_index_df['total_price_index'], 'r')
ax.plot(new_index_df['date_time'], new_index_df['fresh_vegi_index'], 'g')
ax.plot(new_index_df['date_time'], new_index_df['living_price_index'], 'b')
ax.plot(new_index_df['date_time'], new_index_df['food_living_index'], 'y')

# 플롯 제목 및 범례 설정
ax.set_title('일별 지수')
ax.legend(['total_price_index', 'fresh_vegi_index', 'living_price_index', 'food_living_index'])

# 플롯 표시
plt.show()
```
