```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
color_pal = sns.color_palette()
plt.style.use('fivethirtyeight')
```

```python
yearly_area_prod_df = pd.read_excel('2013_2023_onion_area_product.xlsx')

# yearly_area_prod_df에 year, month 컬럼 추가
yearly_area_prod_df = yearly_area_prod_df.rename(columns={'datetime': 'date_time'})

# year, month 컬럼 추가
yearly_area_prod_df['year'] = pd.DatetimeIndex(yearly_area_prod_df['date_time']).year
yearly_area_prod_df['month'] = pd.DatetimeIndex(yearly_area_prod_df['date_time']).month

yearly_area_prod_df = pd.read_excel('2013_2023_onion_area_product.xlsx')

# yearly_area_prod_df에 year, month 컬럼 추가
yearly_area_prod_df = yearly_area_prod_df.rename(columns={'datetime': 'date_time'})

# year, month 컬럼 추가
yearly_area_prod_df['year'] = pd.DatetimeIndex(yearly_area_prod_df['date_time']).year

# 생산량(톤) -> 연단위생산량
yearly_area_prod_df = yearly_area_prod_df.rename(columns={'생산량(톤)': '연단위생산량(톤)'})

yearly_area_prod_df
```

```python
monthly_prod_df = pd.read_excel('2020_2022_onion_product_month.xlsx')

# 컬럼 정리
monthly_prod_df = monthly_prod_df.rename(columns={'datetime': 'date_time'})
monthly_prod_df = monthly_prod_df.drop('품목', axis=1)



# 날짜 범위가 '2020-01-01'에서 '2022-11-01'인 새 데이터 프레임 생성
new_date_range = pd.date_range(start='2020-01-01', end='2022-11-01', name='date_time')
new_withnull_df = pd.DataFrame({'date_time': new_date_range})

# 새 데이터 프레임을 기존 데이터 프레임과 병합
daily_prod_df = pd.merge(new_withnull_df, monthly_prod_df, on='date_time', how='left')

# date_time을 기준으로 데이터 프레임 정렬
daily_prod_df = daily_prod_df.sort_values('date_time').reset_index(drop=True)
daily_prod_df
```

```python
# null 값을 interpolate 써서 채워줌
daily_prod_df = daily_prod_df.interpolate()
daily_prod_df
```

```python
# 월별 값을 일별로 /30 해주기
daily_prod_df['생산량(kg)'] = daily_prod_df['생산량(kg)'] / 30
daily_prod_df
```

```python
# year, month 컬럼 추가
daily_prod_df['year'] = pd.DatetimeIndex(daily_prod_df['date_time']).year
daily_prod_df['month'] = pd.DatetimeIndex(daily_prod_df['date_time']).month

# 생산량(kg) -> 월단위생산량
daily_prod_df = daily_prod_df.rename(columns={'생산량(kg)': '월단위생산량(kg)'})

daily_prod_df
```

```python
#'year'컬럼 기준으로 두 데이터프레임 합침
area_prod_df = pd.merge(yearly_area_prod_df, daily_prod_df, on='year')

# 겹치는 date_time 제거
area_prod_df = area_prod_df.rename(columns={'date_time_y': 'date_time'})
area_prod_df = area_prod_df.drop('date_time_x', axis=1)

# 컬럼 순서 바꾸기
area_prod_df = area_prod_df[['date_time', 'year', '재배면적합계(ha)', '10a당생산량(kg)', '연단위생산량(톤)', 'month', '월단위생산량(kg)']]

# 결과 출력
area_prod_df
```

```python
#### 연도별 재배면적 & 면적당 재배량 및 월별 재배량

# 원하는 너비 및 높이 설정
fig, ax = plt.subplots(figsize=(15,5))

# plot 표시
ax.plot(area_prod_df['date_time'], area_prod_df['월단위생산량(kg)'], 'bo')

# 플롯 제목 및 범례 설정
ax.set_title('월별 지수')
ax.legend(['월단위생산량(kg)'])

# 플롯 표시
plt.show()
```

```python
#### 연도별 재배면적 & 면적당 재배량 및 월별 재배량

# 원하는 너비 및 높이 설정
fig, ax = plt.subplots(figsize=(15,5))

# plot 표시
ax.plot(area_prod_df['date_time'], area_prod_df['재배면적합계(ha)'], 'r')


# 플롯 제목 및 범례 설정
ax.set_title('월별 지수')
ax.legend(['재배면적합계(ha)'])

# 플롯 표시
plt.show()
```

```python
#### 연도별 재배면적 & 면적당 재배량 및 월별 재배량

# 원하는 너비 및 높이 설정
fig, ax = plt.subplots(figsize=(15,5))

# plot 표시
ax.plot(area_prod_df['date_time'], area_prod_df['10a당생산량(kg)'], 'g')


# 플롯 제목 및 범례 설정
ax.set_title('월별 지수')
ax.legend(['10a당생산량(kg)'])

# 플롯 표시
plt.show()
```

```python
# date_time을 인덱스로 설정
area_prod_df = area_prod_df.set_index('date_time')
area_prod_df.index = pd.to_datetime(area_prod_df.index)
area_prod_df
```
