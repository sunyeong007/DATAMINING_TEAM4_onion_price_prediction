```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
color_pal = sns.color_palette()
plt.style.use('fivethirtyeight')
```

```python
weather_df = pd.read_excel('weather_after2013.xlsx')

weather_df
```

```python
# 평균풍속, 평균일조율을 새로운 컬럼으로 추가하기
weather_df['평균풍속(m/s)'] = weather_df[['평균풍속(m/s)_통영', '평균풍속(m/s)_울진', '평균풍속(m/s)_여수']].mean(axis=1)
weather_df['일조율(%)'] = weather_df[['일조율(%)_통영', '일조율(%)_울진', '일조율(%)_여수']].mean(axis=1)


# 일시 통일
weather_df = weather_df.rename(columns={'일시': 'date_time'})

weather_df.head()
```

```python
# 날짜 범위가 '2013-01-01'에서 '2023-05-14'인 새 데이터 프레임 생성
new_date_range = pd.date_range(start='2013-01-01', end='2023-05-14', name='date_time')
new_withnull_df = pd.DataFrame({'date_time': new_date_range})

# 새 데이터 프레임을 기존 데이터 프레임과 병합
weather_df = pd.merge(new_withnull_df, weather_df, on='date_time', how='left')

# date_time을 기준으로 데이터 프레임 정렬
weather_df = weather_df.sort_values('date_time').reset_index(drop=True)
```

```python
weather_df.isnull().sum()
# 원본데이터 null 값 존재하지 않음
```

```python
#### 기온 plot

# 원하는 너비 및 높이 설정
fig, ax = plt.subplots(figsize=(15, 6))

# plot 표시
ax.plot(weather_df['date_time'], weather_df['평균기온(℃)'], 'r')
ax.plot(weather_df['date_time'], weather_df['최고기온(℃)'], 'g')
ax.plot(weather_df['date_time'], weather_df['최저기온(℃)'], 'b')
ax.plot(weather_df['date_time'], weather_df['일교차'], 'y')

# 플롯 제목 및 범례 설정
ax.set_title('기온')
ax.legend(['평균기온(℃)', '최고기온(℃)', '최저기온(℃)', '일교차'])

# 플롯 표시
plt.show()
```

```python
#### 강수량 plot

# 원하는 너비 및 높이 설정
fig, ax = plt.subplots(figsize=(15, 6))

# plot 표시
ax.plot(weather_df['date_time'], weather_df['강수량(mm)'], 'b')

# 플롯 제목 및 범례 설정
ax.set_title('강수량')
ax.legend(['강수량(mm)'])

# 플롯 표시
plt.show()
```

```python
#### 습도 plot

# 원하는 너비 및 높이 설정
fig, ax = plt.subplots(figsize=(15, 6))

# plot 표시
ax.plot(weather_df['date_time'], weather_df['평균습도(%rh)'], 'g')
ax.plot(weather_df['date_time'], weather_df['최저습도(%rh)'], 'b')

# 플롯 제목 및 범례 설정
ax.set_title('습도')
ax.legend(['평균습도(%rh)', '최저습도(%rh)'])

# 플롯 표시
plt.show()
```

```python
#### 일조합&일조율 plot

# 원하는 너비 및 높이 설정
fig, ax = plt.subplots(figsize=(15, 6))

# plot 표시
ax.plot(weather_df['date_time'], weather_df['일조합(hr)'], 'ro')
ax.plot(weather_df['date_time'], weather_df['일조율(%)'], 'go')

# 플롯 제목 및 범례 설정
ax.set_title('일조합&일조율')
ax.legend(['일조합(hr)', '일조율(%)'])

# 플롯 표시
plt.show()
```

```python
#### 평균풍속 plot

# 원하는 너비 및 높이 설정
fig, ax = plt.subplots(figsize=(15, 6))

# plot 표시
ax.plot(weather_df['date_time'], weather_df['평균풍속(m/s)'], 'b')

# 플롯 제목 및 범례 설정
ax.set_title('평균풍속')
ax.legend(['평균풍속(m/s)'])

# 플롯 표시
plt.show()
```

```python
#### 지역별 평균풍속 plot

# 원하는 너비 및 높이 설정
fig, ax = plt.subplots(figsize=(15, 6))

# plot 표시
ax.plot(weather_df['date_time'], weather_df['평균풍속(m/s)_여수'], 'r')
ax.plot(weather_df['date_time'], weather_df['평균풍속(m/s)_통영'], 'b')
ax.plot(weather_df['date_time'], weather_df['평균풍속(m/s)_울진'], 'g')

# 플롯 제목 및 범례 설정
ax.set_title('지역별 평균풍속')
ax.legend(['평균풍속(m/s)_통영', '평균풍속(m/s)_울진', '평균풍속(m/s)_여수'])

# 플롯 표시
plt.show()
```

```python
#### 일조율 plot

# 원하는 너비 및 높이 설정
fig, ax = plt.subplots(figsize=(15, 6))

# plot 표시
ax.plot(weather_df['date_time'], weather_df['일조율(%)'], 'b')

# 플롯 제목 및 범례 설정
ax.set_title('일조율')
ax.legend(['일조율(%)'])

# 플롯 표시
plt.show()
```
