```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
color_pal = sns.color_palette()
plt.style.use('fivethirtyeight')
```

```python
price_df = pd.read_excel('양파가격전처리/onion_price_data.xlsx')
oil_df = pd.read_excel('경유전처리/경유.xlsx')
weather_df = pd.read_excel('기상전처리/weather_df_after2020.xlsx')
quantity_df = pd.read_excel('재배량전처리/new_quantity_area_2020_22.xlsx')
index_df = pd.read_excel('지수전처리/index_data.xlsx')
moist_df = pd.read_excel('토양습도전처리/moist_data.xlsx')
```

```python
# 합쳐야 할 데이터프레임
dataframes = [price_df, oil_df, weather_df, quantity_df, index_df, moist_df]

# date_time 컬럼을 기준으로 데이터프레임 합침
merged_df = dataframes[0]
for df in dataframes[1:]:
    merged_df = pd.merge(merged_df, df, on='date_time', how='inner')

# 합쳐진 데이터프레임
merged_df
```

```python
# date_time을 인덱스로 설정
merged_df = merged_df.set_index('date_time')
merged_df.index = pd.to_datetime(merged_df.index)
merged_df
```

```python
merged_df.to_excel(excel_writer='../2020년이후_1개월단위예측/onion_after2020_combined.xlsx', sheet_name='onion_after2020_combined')
```
