```python
# 6개월 전 평균치 -> 양파 키우는 기간 동안의 기상, 토양, 농기계 경유
 
for i in [5, 11, 23, 35, 47, 59, 71, 83, 95, 107]:
    for col_name in ['자동차용경유', '평균습도(%rh)','강수량(mm)', '일조합(hr)', '평균기온(℃)', 
                     '일조율(%)', '평균풍속(m/s)', '일교차']:   
        ##### 각 행에 1,2,3,4,5,6개월 전의 평균 값 추가하기
        # 새로운 df에 각 월의 평균 값을 구한 후, 값을 한 달씩 뒤로 미룸 -> 전월의 평균값을 나타내도록
        monthly_rainfall_mean2 = df.groupby(pd.Grouper(key='date_time', freq='MS'))[str(col_name)].mean()
        monthly_rainfall_mean2 = monthly_rainfall_mean2.shift(periods=i+2, freq='M').reset_index()

        # 새로운 df에 year, month 컬럼 추가
        monthly_rainfall_mean2['year'] = pd.DatetimeIndex(monthly_rainfall_mean2['date_time']).year
        monthly_rainfall_mean2['month'] = pd.DatetimeIndex(monthly_rainfall_mean2['date_time']).month

        # 새로 생긴 컬럼 이름을 바꿔줌
        monthly_rainfall_mean2 = monthly_rainfall_mean2.rename(columns={str(col_name): str(i+1)+'개월전'+str(col_name)})

        # 새로 만든 컬럼을 연도 및 월을 기준으로 원래 데이터 프레임과 병합
        df = pd.merge(df, monthly_rainfall_mean2, on=['year', 'month'], how='left')

        # 겹치는 date_time 제거
        df = df.rename(columns={'date_time_x': 'date_time'})
        df = df.drop('date_time_y', axis=1)
```

```python
#  6개월, N년 전 평균 -> 상품중품평균가격

for i in [7,13, 25, 37, 49, 61, 73, 85, 97, 109]:
    for col_name in ['상품중품평균가격','소비자물가지수', '신선야채지수', '생활물가지수', '식품생활지수']:   
        ##### 각 행에 1개월, N년 전의 평균 값 추가하기
        # 새로운 df에 각 월의 평균 값을 구한 후, 값을 한 달씩 뒤로 미룸 -> 전월의 평균값을 나타내도록
        monthly_rainfall_mean2 = df.groupby(pd.Grouper(key='date_time', freq='MS'))[str(col_name)].mean()
        monthly_rainfall_mean2 = monthly_rainfall_mean2.shift(periods=i, freq='M').reset_index()

        # 새로운 df에 year, month 컬럼 추가
        monthly_rainfall_mean2['year'] = pd.DatetimeIndex(monthly_rainfall_mean2['date_time']).year
        monthly_rainfall_mean2['month'] = pd.DatetimeIndex(monthly_rainfall_mean2['date_time']).month

        # 새로 생긴 컬럼 이름을 바꿔줌
        monthly_rainfall_mean2 = monthly_rainfall_mean2.rename(columns={str(col_name): str(i-1)+'개월전'+str(col_name)})

        # 새로 만든 컬럼을 연도 및 월을 기준으로 원래 데이터 프레임과 병합
        df = pd.merge(df, monthly_rainfall_mean2, on=['year', 'month'], how='left')

        # 겹치는 date_time 제거
        df = df.rename(columns={'date_time_x': 'date_time'})
        df = df.drop('date_time_y', axis=1)
```
